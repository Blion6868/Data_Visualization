# Census Income Visualization

Data source: https://www.census.gov/topics/income-poverty/income/data/datasets.html

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\bryan\Desktop\adult.csv")
df.head()
```

```python
df.describe()
```

```python
sns.countplot(data=df, x='Gender')
plt.xticks(rotation=0)
plt.show()
```

```python
gender_counts = df['Gender'].value_counts()

gender_percentages = (gender_counts / gender_counts.sum()) * 100

print(gender_percentages)
```

```python
sns.countplot(data=df, x='Income')
plt.xticks(rotation=0)
plt.show()
```

```python
income_counts = df['Income'].value_counts()

income_percentages = (income_counts / income_counts.sum()) * 100

print(income_percentages)
```

```python
sns.countplot(data=df, x='Race')
plt.xticks(rotation=90) 
plt.show()
```

```python
race_counts = df['Race'].value_counts()

race_percentages = (race_counts / race_counts.sum()) * 100

print(race_percentages)
```

```python
sns.countplot(data=df, x='Marital Status')
plt.xticks(rotation=90) 
plt.show()
```

```python
plt.figure(figsize=(18, 8))
sns.histplot(data=df, x='Age', bins=30, kde=True)
plt.xticks(rotation=90)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution Histogram')

plt.show()
```

```python
plt.figure(figsize=(18, 8))
sns.countplot(data=df, x='Age', order=df['Age'].value_counts().index)
plt.xticks(rotation=90)
mean_age = df['Age'].mean()
plt.show()
```

```python
plt.figure(figsize=(18, 8)) 
sns.countplot(data=df, x='Workclass', order=df['Workclass'].value_counts().index)
plt.xticks(rotation=90)
plt.show() 
```

```python
plt.figure(figsize=(18, 8)) 
sns.countplot(data=df, x='Occupation', order=df['Occupation'].value_counts().index)
plt.xticks(rotation=90)
plt.show() 
```

```python
plt.figure(figsize=(18, 8)) 
sns.countplot(data=df, x='Education', order=df['Education'].value_counts().index)
plt.xticks(rotation=90)
plt.show() 
```

```python
plt.figure(figsize=(18, 8))
sns.histplot(data=df, x='Hours per Week', bins=20, kde=True)
plt.xticks(rotation=90)
plt.xlabel('Hours worked per week')
plt.ylabel('Frequency')
plt.title('Hours Worked Per Week Distribution Histogram')

plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(x="Hours per Week", y="Income", data=df, color='blue', linestyle='-', label='Income vs. Hours per Week')
plt.xlabel('Hours per Week')
plt.ylabel('Income')
plt.title('Income vs. Hours per Week')
plt.legend()
plt.show()
```

```python
sns.set_theme(style="darkgrid")
plt.figure(figsize=(18, 8)) 
sns.lineplot(x="Hours per Week",y="Gender",
             data=df)

df['Hours per Week'].mean()
```

```python
sns.set_theme(style="darkgrid")
plt.figure(figsize=(18, 8)) 
sns.lineplot(x="Hours per Week",y="Age",
             data=df)

df['Hours per Week'].mean()
```
